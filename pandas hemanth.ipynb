```python
def comp(ty):
  a=g["Fuel_type"]=="petrol"
  b=g["Company name"]==ty
  c=g["Year"]=="2015"
  d=g[a][b][c]
  d.to_csv("{}.csv".format(ty))
```

```python
d=list(set(g["Company name"]))
for i in d:
    comp(i)
```

```python
def comp(ty):
  a=g["Fuel_type"]=="Diesel"
  b=g["Company name"]==ty
  c=g["Year"]=="2015"
  d=g[a][b][c]
  d.to_csv("{}_{}.csv".format(ty,"Diesel"))
```

```python
d=list(set(g["Company name"]))
for i in d:
    comp(i)
```

```python
def comp(ty):
  yr="2009"
  a=g["Fuel_type"]=="Diesel"
  b=g["Company name"]==ty
  c=g["Year"]=="yr"
  d=g[a][b][c]
  d.to_csv("{}_{}_{}.csv".format(ty,"Diesel",yr))
```

```python
d=list(set(g["Company name"]))
for i in d:
    comp(i)
```

```python
def comp(ty):
  yr="2009"
  yr1="2014"
  a=g["Fuel_type"]=="Diesel"
  b=g["Company name"]==ty
  c=g["Year"]=="yr"
  d=g[a][b][c]
  d.to_csv("{}_{}_{}.csv".format(ty,"Diesel",yr,yr1))
```

```python
d=list(set(g["Company name"]))
for i in d:
    comp(i)
```
